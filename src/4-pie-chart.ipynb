### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.


import matplotlib.pyplot as plt

# Data for pie chart
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Error handling for input validation
if sum(sizes) <= 0:
    raise ValueError("Sizes must be greater than zero.")

    # Create pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['blue', 'red', 'green', 'purple'])

    # Customize the plot
plt.title('Programming Language Popularity')

    # Show the plot
plt.show()
```
